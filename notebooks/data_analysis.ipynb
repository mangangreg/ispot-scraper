```python
subfolders = list(x for x in datapath.glob('*') if x.is_dir())
subfolders
```

```python
dfs = []
for folder in subfolders:
    category = folder.name
    data = [json.load(open(file)) for file in folder.glob('*.json')]
    df = pd.DataFrame(data).set_index('id')
    df['category'] = category
    dfs.append(df)
```

```python
df_full = pd.DataFrame([])
```

```python
df_full.append(dfs[0])
```

```python
len(dfs)
```

```python
df_full = pd.DataFrame()
for df in dfs:
    df_full = df_full.append(df)
    print(df_full.shape[0])
df_full
```
